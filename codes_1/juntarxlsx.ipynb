```python
import pandas as pd
import os

def merge_excel_files(folder_path, output_file):
    # Lista para armazenar os dataframes
    all_data = []
    
    # Itera pelos arquivos na pasta
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.xls'):
            file_path = os.path.join(folder_path, file_name)
            try:
                # Lê cada arquivo Excel
                df = pd.read_excel(file_path, engine='xlrd')  # 'xlrd' é o engine para arquivos .xls
                all_data.append(df)
            except Exception as e:
                print(f"Erro ao processar o arquivo {file_name}: {e}")
    
    # Verifica se há dados para combinar
    if all_data:
        # Concatena todos os dataframes
        merged_data = pd.concat(all_data, ignore_index=True)
        # Salva o resultado em um novo arquivo Excel
        merged_data.to_excel(output_file, index=False, engine='openpyxl')  # Para salvar em .xlsx
        print(f"Arquivo unificado salvo como {output_file}")
    else:
        print("Nenhum arquivo Excel válido foi encontrado.")

# Especifique o caminho da pasta com os arquivos e o nome do arquivo de saída
folder_path = r'C:\Users\User100\Downloads\contas a receber2'
output_file = 'arquivo_unificado.xlsx'  # Alterado para .xlsx

merge_excel_files(folder_path, output_file)
```

```python
import os
import pandas as pd
from win32com.client import Dispatch

def convert_xls_to_xlsx(file_path):
    excel = Dispatch('Excel.Application')
    wb = excel.Workbooks.Open(file_path)
    new_file_path = file_path + "x"
    wb.SaveAs(new_file_path, FileFormat=51)  # 51 é o código para .xlsx
    wb.Close()
    excel.Quit()
    return new_file_path

def merge_excel_files(folder_path, output_file):
    all_data = []
    
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.xls'):
            file_path = os.path.join(folder_path, file_name)
            try:
                # Converte para .xlsx antes de ler
                xlsx_file_path = convert_xls_to_xlsx(file_path)
                df = pd.read_excel(xlsx_file_path)
                all_data.append(df)
            except Exception as e:
                print(f"Erro ao processar o arquivo {file_name}: {e}")
    
    if all_data:
        merged_data = pd.concat(all_data, ignore_index=True)
        merged_data.to_excel(output_file, index=False)
        print(f"Arquivo unificado salvo como {output_file}")
    else:
        print("Nenhum arquivo Excel válido foi encontrado.")

folder_path = r'C:\Users\User100\Downloads\contas a receber2'
output_file = 'arquivo_unificado.xlsx'

merge_excel_files(folder_path, output_file)
```
