```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from googlesearch import search

# Definindo o produto para buscar
produto = "BATOM MATTE( BATOM E BRILHO LABIAL ) FEBELLA"

# Função para extrair o preço do produto de uma página
def get_price_from_page(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"Failed to access {url}")
            return None
        soup = BeautifulSoup(response.content, 'html.parser')

        # Imprimindo o título da página para verificação
        print(f"Accessed {url} - Title: {soup.title.string if soup.title else 'No title'}")

        # Imprimindo uma amostra do HTML para entender a estrutura
        html_sample = soup.prettify()[:1000]  # Imprime os primeiros 1000 caracteres do HTML
        print(html_sample)

        # Tentando encontrar o preço com base em diferentes seletores comuns
        price_selectors = [
            {'name': 'span', 'attrs': {'class': 'price'}},
            {'name': 'span', 'attrs': {'class': 'preco'}},
            {'name': 'span', 'attrs': {'class': 'value'}},
            {'name': 'div', 'attrs': {'class': 'price'}},
            {'name': 'div', 'attrs': {'class': 'preco'}},
            {'name': 'div', 'attrs': {'class': 'value'}}
        ]
        
        for selector in price_selectors:
            price_tag = soup.find(selector['name'], selector['attrs'])
            if price_tag:
                print(f"Found price using selector {selector}: {price_tag.text.strip()}")
                return price_tag.text.strip()

    except Exception as e:
        print(f"Error accessing {url}: {e}")

    return None

# Lista para armazenar os dados
data = []

# Usando o Google para buscar o produto
search_results = search(produto, num_results=10)

# Iterando sobre os resultados de busca e coletando os preços
for url in search_results:
    price = get_price_from_page(url)
    if price:
        data.append({'site': url, 'price': price})

# Convertendo os dados para um DataFrame do pandas
df = pd.DataFrame(data)

# Exibindo a tabela
print(df)

# Salvar a tabela em um arquivo CSV (opcional)
df.to_csv('precos_batom_matte.csv', index=False)
```
