```python
#importações
import os
from PIL import Image
```

```python
#criando as funções
def get_image_info(image_path):
    with Image.open(image_path) as img:
        width, height = img.size
        dimension = f"{width} x {height}"
    size = os.path.getsize(image_path)
    size_kb = size // 1024  # Convertendo bytes para kilobytes
    return dimension, size_kb, width, height

def print_images_info(folder_path):
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('png', 'jpg', 'jpeg', 'gif', 'bmp', 'tiff')):
            image_path = os.path.join(folder_path, filename)
            dimension, size_kb, width, height = get_image_info(image_path)
            print(f"Nome da foto: {filename}")
            print(f"Dimensão: {dimension}")
            print(f"Tamanho: {size_kb} KB")
            if size_kb > 350 or (width < 900 and height < 900):
                print("**Especificações destacadas**")
            print()
```
