```python
import pandas as pd
import openpyxl
import requests
import os
from PIL import Image
from io import BytesIO
```

```python
df = pd.read_excel(r'C:\Users\User100\Documents\Arquivos Renan\from drive\Codes Backup\codes_3\imagens_ready.xlsx')
```

```python
display(df)
```

```python
pasta_imagens = "imagens_3"
os.makedirs(pasta_imagens, exist_ok=True)
```

```python

# Iterar pelas linhas da planilha e baixar as imagens
for index, row in df.iterrows():
    codigo_barras = str(row["Codigo de Barras"])  # Convertendo argumento para string
    link_imagem = row["imagem_1"]

    if pd.notna(link_imagem):  # Verifica se o link não é nulo
        try:
            resposta = requests.get(link_imagem, timeout=10)  # requisição da imagem
            if resposta.status_code == 200:
                # Salvar a imagem com o código de barras como nome em PNG
                caminho_arquivo = os.path.join(pasta_imagens, f"{codigo_barras}.png")
                with open(caminho_arquivo, "wb") as arquivo:
                    arquivo.write(resposta.content)
                print(f"Imagem {codigo_barras} salva com sucesso.")
            else:
                print(f"Erro ao baixar a imagem {codigo_barras}. Status: {resposta.status_code}")
        except Exception as e:
            print(f"Erro ao processar a imagem {codigo_barras}: {e}")
    else:
        print(f"Sem link de imagem para o SKU na linha {index + 1}.")
```

```python
# Largura fixa desejada
largura_desejada = 600

# Iterar pelas linhas da planilha e baixar as imagens
for index, row in df.iterrows():
    codigo_barras = str(row["Codigo de Barras"])  # Converter para string
    link_imagem = row["imagem_1"]

    if pd.notna(link_imagem):  # Verificar se há link
        try:
            resposta = requests.get(link_imagem, timeout=10)  # Baixar a imagem
            if resposta.status_code == 200:
                # Abrir a imagem
                imagem = Image.open(BytesIO(resposta.content))

                # Converter para RGB caso a imagem seja PNG ou tenha transparência
                if imagem.mode != "RGB":
                    imagem = imagem.convert("RGB")

                # Redimensionar mantendo a proporção
                proporcao = largura_desejada / float(imagem.width)
                altura_nova = int(imagem.height * proporcao)
                imagem = imagem.resize((largura_desejada, altura_nova), Image.LANCZOS)

                # Salvar a imagem como JPG
                caminho_arquivo = os.path.join(pasta_imagens, f"{codigo_barras}.jpg")
                imagem.save(caminho_arquivo, format="JPEG", quality=90)  # Qualidade ajustada para 90

                print(f"Imagem {codigo_barras} salva com sucesso ({largura_desejada}x{altura_nova}).")
            else:
                print(f"Erro ao baixar a imagem {codigo_barras}. Status: {resposta.status_code}")
        except Exception as e:
            print(f"Erro ao processar a imagem {codigo_barras}: {e}")
    else:
        print(f"Sem link de imagem para o SKU na linha {index + 1}.")
```
