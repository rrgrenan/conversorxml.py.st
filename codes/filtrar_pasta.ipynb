```python
import os
import shutil

# Caminho da pasta de origem com todas as imagens
caminho_origem = r"C:\Users\55719\OneDrive\Documentos\Renan\Engenharia\Programação\codigos\imagens_2"
# Caminho da pasta de destino onde as imagens filtradas serão copiadas
caminho_destino = r"C:\Users\55719\OneDrive\Documentos\Renan\Engenharia\Programação\codigos\imagem_franc"
# Lista de SKUs dos produtos da marca que você quer manipular
skus_validos = [
    'BASE MATE C-01', 'BASE MATE C-02', 'BASE MATE C-03', 'BASE MATE C-04',
    'BASE MATTE E-01', 'BASE MATE M-01', 'BASE MATE M-02', 'BASE MATE M-03',
    'BASE MATE M-04', 'BATOM CLARICE', 'BATOM DELICATE', 'BATOM SENSUALIZANI',
    'BATOM XOXO', 'BATOM LIQ. ANDROMEDA', 'BATOM LIQ. PEGASUS',
    'CANETA DELINEADORA PULLLPEN', 'CHOCOCHILLI', 'CORRETIVO M-01', 
    'CORRETIVO E-01', 'CORRETIVO M-02', 'FIXADOR POWER', 'GLOSS AUMENTO LABIAL',
    'GLOSSY GLOSSIP GIRL', 'GLOSS LABIAL MARTE', 'GLOSS LABIAL VENUS',
    'GLOSS LABIAL TATTOO', 'OIL MERCURIO', 'GREENCHILLI', 'GLOSSLICIOUS',
    'LAPISEIRA PULLPENCIL', 'MÁSCARA MARAVICHERRY', 'MÁSC. CÍLIOS NETUNO',
    'MÁSCARA PLUTÃO', 'PÓ FACIAL SOLTO PLUSH 1', 'PÓ FACIAL SOLTO PLUSH 2',
    'PÓ SOLTO PLUSH 3', 'PÓ FACIAL ROSA/PINK', 'PRIMER FACIAL', 'SERUM C10',
    'STICK TINT BAMBOO', 'STICK TINT PINK', 'STICK TINT WINE'
]

# Converter todos os SKUs válidos para minúsculas e remover espaços extras
skus_validos = [sku.strip().lower() for sku in skus_validos]

# Certifique-se de que a pasta de destino exista
os.makedirs(caminho_destino, exist_ok=True)

# Lista todas as imagens na pasta de origem
imagens = os.listdir(caminho_origem)

for imagem in imagens:
    # Extrai o SKU do nome do arquivo (assumindo que o nome do arquivo é o SKU)
    sku = os.path.splitext(imagem)[0].strip().lower()
    print(f"Verificando imagem: {imagem} - SKU: {sku}")  # Linha de debug
    if sku in skus_validos:
        # Copia a imagem para a pasta de destino
        print(f"Copiando imagem: {imagem}")  # Linha de debug
        shutil.copy(os.path.join(caminho_origem, imagem), caminho_destino)
    else:
        print(f"SKU não encontrado na lista: {sku}")  # Linha de debug
```
