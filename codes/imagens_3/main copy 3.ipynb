```python
import requests
from PIL import Image
import io
import pandas as pd
from rembg import remove  # Certifique-se de que voc tem o mdulo rembg instalado
import time

planilha_2 = pd.read_excel("NAO SALVAS.xlsx")

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

def sanitize_filename(filename):
    return filename.replace('/', '_').replace('\\', '_')

def download_image(url, retries=5, delay=2):
    for i in range(retries):
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()  # Verifica se a requisio foi bem-sucedida
            return response.content
        except requests.exceptions.RequestException as e:
            print(f"Tentativa {i+1}/{retries} falhou: {e}")
            time.sleep(delay)
    raise Exception("Falha ao baixar a imagem aps mltiplas tentativas")

for index, row in planilha_2.iterrows():
    nome_produto = row['COD']
    url_imagem = row['URL']
    
    # Verificar se a URL  vlida
    if pd.notna(url_imagem):
        try:
            # Fazer o download da imagem
            image_data = download_image(url_imagem)
            
            # Criar um objeto de imagem a partir da imagem baixada
            image = Image.open(io.BytesIO(image_data))
            
            # Remover o fundo da imagem (supondo que voc esteja usando uma funo chamada remove)
            image_with_transparent_background = remove(image)
            
            # Sanitizar o nome do arquivo
            nome_arquivo = sanitize_filename(f"{nome_produto}.png")
            image_with_transparent_background.save(nome_arquivo, format="PNG")
            
            # Resultado:
            print(f"Imagem salva como {nome_arquivo}")
        except Exception as e:
            print(f"No foi possvel baixar a imagem para o produto '{nome_produto}': {e}")
    else:
        print(f"URL invlida para o produto '{nome_produto}'")
```
