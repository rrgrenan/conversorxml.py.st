```python
import pandas as pd
import os
import glob
from PIL import Image
```

```python
caminho_pasta = r"C:\Users\USER\Downloads\novaspng"
```

```python
# Verificar se a pasta existe
if not os.path.exists(caminho_pasta):
    print(f"A pasta '{caminho_pasta}' não existe.")
else:
    # Listar todos os arquivos de imagem na pasta (considerando várias extensões de imagem)
    extensoes_imagem = ['*.jpg', '*.jpeg', '*.png', '*.gif', '*.bmp', '*.tiff']

    # Criar uma lista para armazenar os nomes dos arquivos de imagem
    arquivos_imagens = []

    # Iterar por cada extensão e adicionar os arquivos encontrados à lista
    for extensao in extensoes_imagem:
        arquivos_encontrados = glob.glob(os.path.join(caminho_pasta, extensao))
        print(f"Arquivos encontrados para a extensão {extensao}: {arquivos_encontrados}")
        arquivos_imagens.extend(arquivos_encontrados)

    # Verificar se encontramos algum arquivo de imagem
    if not arquivos_imagens:
        print("Nenhum arquivo de imagem encontrado na pasta especificada.")
    else:
        # Exibir os nomes dos arquivos de imagem
        for arquivo in arquivos_imagens:
            print(os.path.basename(arquivo))
```

```python


# Caminho para a planilha Excel
caminho_planilha = 'nova_extracao.xlsx'

# Carregar a planilha
df = pd.read_excel(caminho_planilha)

```

```python
print(df)
```

```python
# Supondo que os nomes estejam na primeira coluna (index 0)
nomes = df.iloc[:, 0].tolist()

# Listar os arquivos de imagem na pasta (supondo que sejam PNGs)
arquivos_imagem = [f for f in os.listdir(caminho_pasta) if f.endswith('.png')]

# Ordenar os arquivos de imagem pela data de criação
arquivos_imagem.sort(key=lambda x: os.path.getctime(os.path.join(caminho_pasta, x)))

# Reverter a lista para que a ordem seja do mais antigo ao mais novo
arquivos_imagem.reverse()

# Verificar se o número de nomes corresponde ao número de arquivos de imagem
if len(nomes) != len(arquivos_imagem):
    raise ValueError("O número de nomes na planilha não corresponde ao número de arquivos de imagem")

# Iterar sobre os nomes e arquivos de imagem
for i, nome in enumerate(nomes):
    # Caminho atual do arquivo de imagem
    caminho_atual = os.path.join(caminho_pasta, arquivos_imagem[i])
    
    # Novo caminho com o nome do arquivo renomeado
    novo_nome = f"{nome}.png"
    novo_caminho = os.path.join(caminho_pasta, novo_nome)
    
    # Renomear o arquivo
    os.rename(caminho_atual, novo_caminho)
    
    print(f"Renomeado {arquivos_imagem[i]} para {novo_nome}")

print("Renomeação concluída.")
```
